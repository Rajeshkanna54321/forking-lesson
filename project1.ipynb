```python
import pandas as pd

# Load TSV and rename columns
df = pd.read_csv("/content/hate_speech_detection/available_metahate.tsv", sep="\t")
df.columns = [c.strip().lower() for c in df.columns]
if 'label1' in df.columns:
    df = df.rename(columns={'label1': 'label'})
elif 'labels' in df.columns:
    df = df.rename(columns={'labels': 'label'})
if 'tweet' in df.columns:
    df = df.rename(columns={'tweet': 'text'})
elif 'comment' in df.columns:
    df = df.rename(columns={'comment': 'text'})
elif 'sentence' in df.columns:
    df = df.rename(columns={'sentence': 'text'})

# Save in the format expected by metahate_bart.py
df.to_csv("/content/hate_speech_detection/finetuned/metahate.csv", index=False)
print("✅ Saved as metahate.csv with columns:", df.columns.tolist())
```
